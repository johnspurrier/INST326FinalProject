```python
import pandas as pd

df_skaters = pd.read_csv("speed_skaters.csv")
df_skaters.head()
```

```python
df1 = pd.read_csv("mens1500_2006.csv")
df2 = pd.read_csv("mens1500_2010.csv")
df3 = pd.read_csv("mens1500_2014.csv")
df4 = pd.read_csv("mens1500_2018.csv")
```

```python
df_combined = pd.concat([df1,df2,df3,df4])
df = pd.merge(df_skaters, df_combined, left_on="Name",right_on="Athlete",how="inner").drop(["Athlete"],axis=1)
df.head()
```

```python
df[(df["Height"] > 1.80) & (df["Weight"] > 80)]
```

```python
df_birth_year = df.groupby("Birth year")
mean_weight = df_birth_year["Weight"].mean()
mean_weight
```
