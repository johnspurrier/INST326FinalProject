A DataFrame is a two dimensional data with an index, columns are a series (each has a label and position), each row has a label (from the index) and a position.

```python
import pandas as pd

df = pd.read_csv("DisneylandReviews.csv", engine = 'python')
df.head()
```

```python
df[(df["Rating"] > 4) & (df["Year_Month"] >= "2019-4")]
```

For the boolean indexing I chose to pull all of the reviews that were rated higher than 4 and came during of after April of 2019 by using the & operator. This works because 

To select a column from a DataFrame using the column label, you have to use the [] with the label of the column in it. You can do this with as many columns as you want, they just need to be separated by commas.

```python
df[["Rating", "Year_Month", "Reviewer_Location"]]
```
